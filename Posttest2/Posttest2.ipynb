```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('laptop_price.csv', encoding="ISO-8859-1")
df.head()
```

```python
# Rata-rata dari Label Price_euros berdasarkan Company (Merknya)
rerata = df.groupby("Company")["Price_euros"].mean()
rerata
```

```python
# Rata-rata dari Label Price_euros berdasarkan Company (Merknya) dalam bentuk Diagram Batang
x = rerata.index
y = rerata.values
plt.figure(figsize=(18,6))
plt.bar(x, y, color=['indigo'])
plt.title('Rata-rata Harga Laptop berdasarkan Merk')
plt.xlabel('Merk Laptop')
plt.ylabel('Harga Laptop')
plt.show()
```

```python
# Standar Deviasi dari Label Price_euros berdasarkan Company (Merknya)
stdeviasi = df.groupby("Company")["Price_euros"].std()
stdeviasi
```

```python
# Standar Deviasi dari Label Price_euros berdasarkan Company (Merknya) dalam bentuk Diagram Batang
x = stdeviasi.index
y = stdeviasi.values
plt.figure(figsize=(18,6))
plt.bar(x, y, color=['skyblue'])
plt.title('Persebaran Harga Laptop berdasarkan Merk')
plt.xlabel('Merk Laptop')
plt.ylabel('Harga Laptop')
plt.show()
```

```python
# Nilai minimum dari Label Price_euros berdasarkan Company (Merknya)
minprice = df.groupby("Company")["Price_euros"].min()
minprice
```

```python
# Nilai maksimum dari Label Price_euros berdasarkan Company (Merknya)
maxprice = df.groupby("Company")["Price_euros"].max()
maxprice
```

```python
# Q1, Q2, Q3
q1 = df["Price_euros"].quantile(0.25)  # Kuartil 1
q2 = df["Price_euros"].quantile(0.50)  # Kuartil 2 (median)
q3 = df["Price_euros"].quantile(0.75)  # Kuartil 3

print("Q1:", q1)
print("Q2 (Median):", q2)
print("Q3:", q3)
```

```python
# Kuartil dalam bentuk boxplot
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15,6))
sns.boxplot(x=df['Price_euros'], color="skyblue")

plt.title("Boxplot Harga Laptop", fontsize=14)
plt.xlabel("Harga (Euro)")
plt.show()
```

```python
# Menampilkan 10 record pertama
numeric_df = df.select_dtypes(include=['int64', 'float64']) # Biar yang masuk cuman numerik
numeric_df.head(10)
```

```python
# Variable penampung jumlah record
record = df.shape[0]
# Variable penampung jumlah Attribute
attribute = df.shape[1]
# Variable penampung nama Attribute
nama_attribute = df.columns.tolist()
# Variable penampung jumlah Attribute tipe numerik
atribut_numerik = df.select_dtypes(include=['int64', 'float64']).shape[1]

print("Jumlah record:", record)
print("Jumlah attribute:", attribute)
print("Attribute:", nama_attribute)
print("Jumlah attribute dengan tipe numerik:", atribut_numerik)
```

```python
# Ini untuk select yang tipenya object
object_cols = df.select_dtypes(include=['object']).columns

# Untuk hitung label unik
for col in object_cols:
    print(f"{col} -> {df[col].nunique()} label unik")
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df['Price_euros'], bins=30, kde=True, color="skyblue")
plt.title("Distribusi Harga Laptop", fontsize=14)
plt.xlabel("Harga (Euro)")
plt.ylabel("Jumlah Laptop")
plt.show()
```

```python
plt.figure(figsize=(6,4))
corr = df.select_dtypes(include=['number']).corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Heatmap Korelasi Atribut Numerik")
plt.show()
```
